```python
import numpy as np
from paddedDiffraction import fft2Pad
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def fork(k, m, x, y, hard=False):
    val = np.sin(k * x + m * np.arctan2(y, x))
    if hard: val = np.sign(val)    
    #Make grating go from 0 to 1 instead of -1 to 1.
    val = (val + 1.0) / 2.0
    return val

def fork_arr(k, m, X, Y, hard=False, thin=False, cutoff=0.95):
    arr = fork(k, m, X, Y, hard=hard)
    if thin:
        arr = (arr >= cutoff * np.max(arr))
    return arr

def rotate(x0, y0, angle, measure='radians'):
    #rotates tuple (x0, y0) counterclockwise by a given angle.
    angle = float(angle)
    if measure == 'degrees':
        angle = angle * np.pi / 180.0
    x1 = np.cos(angle) * x0 - np.sin(angle) * y0
    x2 = np.sin(angle) * x0 + np.cos(angle) * y0
    return x1, x2

def amp_to_phase(amp):
    phase = np.exp(np.pi * 1j * amp)
    return phase

def first_order_rad(k, padFactor=0.0):
    pitch = k / (2.0 * np.pi)
    return (np.pi**2 / 8.0) * (padFactor + 1.0) * pitch

def LG(m, w0, wavelength, x, y, z=0):
    zR = np.pi * w0**2 / wavelength
    psiz = (2.0 * m + 1.0) * np.arctan(z / zR)
    wz = w0 * np.sqrt(1.0 + (z / zR)**2.0)
    if not z == 0:
        Rz = z * (1.0 + (zR / z)**2.0)
    k = 2.0 * np.pi / wavelength
    
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    C_ml = np.sqrt(2.0 / (np.pi * np.math.factorial(np.abs(m)))) #coefficient for LG beam with l = 0
    amp = C_ml * (1.0 / wz) * (np.sqrt(2) * rho / wz)**np.abs(m) * np.exp(-rho**2.0 / wz**2.0)
    phase =   np.exp(-1j * m * phi) * np.exp(-1j * k * z) * np.exp(1j * psiz)
    if not z == 0:
        phase = phase * np.exp(-1j * k * rho**2 / (2.0 * Rz))
    
    return amp * phase

def forkList(mMax, mMin=0, k=4.0, hard=False, thin=False, invert=False, cutoff=0.95, mytype='amp'):
    forkList = [X for item in range(mMin, mMax + 1)]
    for i in range(len(forkList)):
        X1, Y1 = rotate(X, Y, (90.0 / (len(forkList) - 1)) * i, measure = 'degrees')
        if not thin:
            forkList[i] = fork(k, mMin + i, X1, Y1, hard=hard)
        else:
            forkList[i] = fork_arr(k, mMin + i, X1, Y1, hard=hard, thin=thin, cutoff=cutoff)
        if mytype == 'phase':
            forkList[i] = np.exp(1j * np.pi * forkList[i])
        if invert:
            forkList[i] = 1 - forkList[i]
        forkList[i] = circ * forkList[i]
    return forkList

def prodFork(mMax, mMin=0, k=4.0, hard=False, thin=False, invert=False, cutoff=0.95, mytype='amp'):
    prod1 = np.prod(forkList(mMax, mMin, k, hard=hard, thin=thin, invert=invert, cutoff=cutoff), axis = 0)
    if mytype == 'phase':
        prod1 = np.exp(1j * np.pi * prod1)
    return prod1

def sumFork(mMax, mMin=0, k=4.0, hard=False, thin=False, invert=False, cutoff=0.95, mytype='amp'):
    sum1 = np.sum(forkList(mMax, mMin, k, hard=hard, thin=thin, invert=invert, cutoff=cutoff, mytype = mytype), axis = 0)
    if mytype == 'phase':
        sum1 = np.exp(1j * np.pi * sum1)
    return sum1

def displayTwo(arr1, arr2, cmap1='gray', cmap2='gray'):

    fig, (ax0, ax1, ax2) = plt.subplots(1, 3, figsize=(15, 5), sharex=True, sharey=True)
    plt.tight_layout()

    ax0.imshow(arr1, cmap=cmap1)
    ax0.set_axis_off()
    ax0.set_adjustable('box-forced')

    ax1.imshow(arr2, cmap=cmap2)
    ax1.set_axis_off()
    ax1.set_adjustable('box-forced')

    ax2.set_axis_off()
    ax2.set_adjustable('box-forced')
    plt.show()
```

## Multiple modes, and offsets

```python
x = np.linspace(-4 * np.pi, 4 * np.pi, 512)
y = np.linspace(-4 * np.pi, 4 * np.pi, 512)
X, Y = np.meshgrid(x,y)

m_1 = 1
m_2 = 5
w0 = 0.4 * np.max(x)
wavelength = np.pi / 2.0 * .004 * 100.0
x_offset_1 = 0.0 * np.max(x)
y_offset_1 = 0.0 * np.max(y)
x_offset_2 = 0.0 * np.max(x)
y_offset_2 = 0.0 * np.max(y)

circ = X**2 + Y**2 <= (np.max(x))**2
inner_circ = X**2 + Y**2 >= (np.max(x) / (2 * np.pi))**2
```

```python
my_fork = fork_arr(4.0, 1.0, X, Y, hard=False, thin=True, cutoff=0.95) * circ
my_LG_1 = LG(m_1, w0, wavelength, X - x_offset_1, Y - y_offset_1)
my_LG_2 = LG(m_2, w0, wavelength, X - x_offset_2, Y - y_offset_2)
my_diff = fft2Pad(my_fork * (my_LG_1 + my_LG_2), padFactor=2.0, norm=False)

displayTwo(np.abs(my_fork * (my_LG_1 + my_LG_2))**2.0, np.abs(my_diff)**2.0 * (np.abs(Y) <= np.max(Y) / 1.0))
```

```python
my_fork = prodFork(5, k=16.0, hard=True, thin=True, invert=True, cutoff=0.95)
my_LG_0 = LG(0, w0, wavelength, X - x_offset_1, Y - y_offset_1)
my_LG_1 = LG(1, w0, wavelength, X - x_offset_1, Y - y_offset_1)
my_LG_2 = LG(2, w0, wavelength, X - x_offset_2, Y - y_offset_2)
my_LG_3 = LG(3, w0, wavelength, X - x_offset_1, Y - y_offset_1)
my_LG_4 = LG(4, w0, wavelength, X - x_offset_2, Y - y_offset_2)
my_LG_5 = LG(5, w0, wavelength, X - x_offset_2, Y - y_offset_2)

my_LG = my_LG_1 + my_LG_3 + my_LG_4
my_diff = fft2Pad(my_fork * my_LG, padFactor=2.0, norm=False)

displayTwo(np.abs(my_fork * my_LG)**2.0, np.abs(my_diff)**2.0 * (np.abs(Y) <= np.max(Y) / 1.0))
```

Note above that, with no offsets, the super gratings I've made does a better job of showing multiple OAM better than does the regular straight grating. If you start putting more than 2 modes in there, though, it starts getting screwed up.

Let's try putting on a mask to isolate the spots where OAM modes should show up.

```python
#masks
inner = X**2 + Y**2 >= (.75 * np.max(x))**2 # chose .75 manually for k = 16
outer = X**2 + Y**2 <= (.76 * np.max(x))**2 # chose .76 manually for k = 16
lat0 = (np.abs((rotate(X, Y, np.pi / 2.0 * (0.0 / 5.0)))[1]) <= .007 * np.max(x))
lat1 = (np.abs((rotate(X, Y, np.pi / 2.0 * (1.0 / 5.0)))[1]) <= .007 * np.max(x))
lat2 = (np.abs((rotate(X, Y, np.pi / 2.0 * (2.0 / 5.0)))[1]) <= .007 * np.max(x))
lat3 = (np.abs((rotate(X, Y, np.pi / 2.0 * (3.0 / 5.0)))[1]) <= .007 * np.max(x))
lat4 = (np.abs((rotate(X, Y, np.pi / 2.0 * (4.0 / 5.0)))[1]) <= .007 * np.max(x))
lat5 = (np.abs((rotate(X, Y, np.pi / 2.0 * (5.0 / 5.0)))[1]) <= .007 * np.max(x))
mask = outer * inner * (lat0 + lat1 + lat2 + lat3 + lat4 + lat5)

displayTwo(mask, np.abs(my_diff)**2.0 * mask)
#displayTwo(np.abs(my_fork * my_LG)**2.0, np.abs(my_diff)**2.0 * inner * outer * lat0)
```

The mask seems to work quite well! Note that these wouldn't even have to be physical masks - they could be added via software as a filter while looking at data, provided that the radial distance of the first diffracted order from the center is known.

Also note, regarding the masks, that smaller cutoffs will cut off part of the beams that are wanted as well. E.g., if the only beam is LG1, and the masks are as stringent as above, less OAM 1 will be quantified than is appropriate.

```python
#masks
circ_rad = 0.0035
circ_mask = ((X - first_order_rad(16.0, padFactor=2))**2 + Y**2 <= circ_rad)
circ0 = circ_mask
X1, Y1 = rotate(X, Y, (1.0 / 5.0) * (np.pi / 2.0));
circ1 = ((X1 - first_order_rad(16.0, padFactor=2))**2 + Y1**2 <= circ_rad)
X1, Y1 = rotate(X, Y, (2.0 / 5.0) * (np.pi / 2.0));
circ2 = ((X1 - first_order_rad(16.0, padFactor=2))**2 + Y1**2 <= circ_rad)
X1, Y1 = rotate(X, Y, (3.0 / 5.0) * (np.pi / 2.0));
circ3 = ((X1 - first_order_rad(16.0, padFactor=2))**2 + Y1**2 <= circ_rad)
X1, Y1 = rotate(X, Y, (4.0 / 5.0) * (np.pi / 2.0));
circ4 = ((X1 - first_order_rad(16.0, padFactor=2))**2 + Y1**2 <= circ_rad)
X1, Y1 = rotate(X, Y, (5.0 / 5.0) * (np.pi / 2.0));
circ5 = ((X1 - first_order_rad(16.0, padFactor=2))**2 + Y1**2 <= circ_rad)

mask_2 = circ0 + circ1 + circ2 + circ3 + circ4 + circ5

displayTwo(mask_2, np.abs(my_diff)**2.0 * mask_2)
#displayTwo(np.abs(my_fork * my_LG)**2.0, np.abs(my_diff)**2.0 * inner * outer * lat0)
```
